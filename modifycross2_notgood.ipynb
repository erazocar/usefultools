```
with open("mytext2.txt") as f:
  content = f.readlines()
  res = ''
  for i in itertools.islice(content, 0, len(content), 2), itertools.islice(content,1,len(content),2):
    res += ' '.join(i)
    #res += '\n'
```

The variable "res" contains now the correct values for each of the variables included per each cross section.
```
l = res.split('\n')
```

```
arr = pd.Series(l)
a = arr[:math.floor(arr.size / 2)-1]
b = arr[math.floor(arr.size / 2):]

a.reset_index()
b.reset_index()
```

```
df = pd.concat([a, b],axis = 1)
```
